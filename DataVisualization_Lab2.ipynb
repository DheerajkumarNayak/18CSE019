```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df=pd.read_csv(r'https://raw.githubusercontent.com/DheerajkumarNayak/18CSE019/main/Bank_churn_modelling.csv')
```

```
df.head()
```

```
df.duplicated().sum() ## cheaking for duplicate value.
```

```
df.isnull().sum() # cheaking for missing value.
```

```
df.drop(['Surname','RowNumber','CustomerId'],axis=1,inplace=True) # Deleting the coloums which are not reqired. 
df.columns
```

```
df.info()
```

```
df.shape
```

## ***Scatter*** **Plot**

```
plt.figure(figsize=(18,10))
plt.scatter(df['Age'],df['CreditScore'],c='green')
plt.xlabel('Age(months)')
plt.ylabel('CreditScore')
plt.show()
```

```
#creditScore v/s age v/s exited
plt.figure(figsize=(12,5))
sns.scatterplot(x='CreditScore',y='Age',hue='Exited',data=df)
plt.show()
```

## **Histogram**

```
plt.hist(df['CreditScore'],edgecolor='white',bins=5)
plt.title('Histogram of CreditScore')
plt.xlabel('Range')
plt.ylabel('CreditScore')
plt.show()
```

## **Regration plot**

```
plt.figure(figsize=(12,5))
sns.set(style='darkgrid')
sns.regplot(x=df['Age'],y=df['Balance'])
```

```
plt.figure(figsize=(18,15))
sns.set(style='darkgrid')
sns.regplot(x=df['CreditScore'],y=df['Balance'])
```

```
plt.figure(figsize=(12,5))
sns.set(style='darkgrid')
sns.regplot(x=df['Age'],y=df['Balance'],marker='*',fit_reg=False) #to replace the dots with Stars and to remove the Regration line.
```

## **Distplot**

```
sns.distplot(df['Age'])
```

```
sns.distplot(df['Age'],kde=False)  #To remove the kernal density estimater(kde)
```

```
# CreaditScore v/s Exited
plt.figure(figsize=(12,5))
sns.distplot(df.CreditScore[df.Exited==0])
sns.distplot(df.CreditScore[df.Exited==1])
plt.legend(['0','1'])
plt.show()
```

```
# Age v/s Exited
plt.figure(figsize=(12,5))
sns.distplot(df.Age[df.Exited==0])
sns.distplot(df.Age[df.Exited==1])
plt.legend(['0','1'])
plt.show()
```

## **Count Plot**

```
plt.figure(figsize=(6,3))
sns.countplot(df.Geography)
plt.show()
```

```
# Geography v/s Exited
plt.figure(figsize=(6,3))
sns.countplot(df.Geography[df.Exited==1])
plt.show()
```

```
plt.figure(figsize=(6,3))
sns.countplot(df.Gender)
plt.show()
plt.figure(figsize=(6,3))
sns.countplot(df.Gender[df.Exited==1]) # Shows the most existed gender
plt.show()
```

# Box Plot

```
sns.boxplot(y='Age',data=df)
```

# Point plot 

```
plt.figure(figsize=(12,5))
sns.pointplot(x='Geography',y='Age',hue='Exited',data=df)
plt.show()
```
